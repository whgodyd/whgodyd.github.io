## NumPy 특징

* Numerical Python의 약자
* 고성능 과학 계산용 패키지로 강력한 N차원 배열 객체

### 배열 생성 및 초기화

* `zeros()`: 모든 요소를 0으로 초기화

```
#10개의 요소를 0으로 초기화한 요소로 만들어줘
np.zeros(10)
```

```
np.zeros(10)
```

* `ones()`: 모든 요소를 1로 초기화

```
# 2차원 배열로 1로 초기화해줘
np.ones((3, 3))
```

```
np.ones((3,3,3))
```

* `full()`: 모든 요소를 지정한 값으로 초기화

```
# 3*3 짜리 특정값으로 채워줘 1.23
np.full((3, 3), 1.23)
```

```
np.full((3,3),1.23)
```

* `eye()`: 단위행렬(identity matrix) 생성
  + 주대각선의 원소가 모두 1이고 나머지 원소는 모두 0인 정사각 행렬, 괄호 안에 n 넣기

```
np.eye(3)
```

```
np.eye(3)
```

* `tri()`: 삼각행렬 생성

```
np.tri(3)
```

```
np.tri(5)
```

* `empty()`: 초기화되지 않은 배열 생성
  + 초기화가 없어서 배열 생성비용 저렴하고 빠름
  + 초기화되지 않아서 기존 메모리 위치에 존재하는 값이 있음(어떤 값이 올지 모름)

```
np.empty(10)
```

```
np.empty(10)
```

* `_like()`: 지정된 배열과 shape가 같은 행렬 생성
  + `np.zeros_like()`
  + `np.ones_like()`
  + `np.full_like()`
  + `np.empty_like()`

```
np.zeros_like(a1)
```

```
print(a2)
np.ones_like(a2)
```

```
print(a3)
np.full_like(a3,10)
```
